```python
# import necessary libraries

from bs4 import BeautifulSoup

import requests

import pandas as pd
```

```python
books = []                                                   #create an empty list

for i in range(1,51):                                        # Loop through the 50 pagesand repeat the same process
    url = f"https://books.toscrape.com/catalogue/page-{i}.html"

    link = requests.get(url)                                  # attach link to a variable

    print(link)

    soup = BeautifulSoup(link.text, 'html')                   # using BeautifulSoup to transform the data into html form

    ol = soup.find('ol')                                      # extrat the order list 'ol'from the data and assign to a variable

    articles = ol.find_all('article', class_='product_pod')   # the variables we need are all inside the 'article' tag under 'ol'



    for article in articles:
        image = article.find('img')                            #<img 
        title = image.attrs['alt']                             #title is embedded inside alt
        star = article.find('p')                               # star rating is within 'class' which is inside <p tag
        star = star['class'][1]                                #'1' to extract only index 1=(2)
        price = article.find('p', class_='price_color').text   #price is inside the <p tag under the class name which is 'price color' and the .text to extract the text only
        price = price[1:]
        price = float(price[1:])                               # 1: to extract only the numbers and convert to a float type
        books.append([title,price,star])                       #.append to add values into the list we created under the 'book' variable
```

```python
df = pd.DataFrame(books, columns=['Title', 'Price', 'Star Rating'])
```
