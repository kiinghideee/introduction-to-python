```python
from bs4 import BeautifulSoup

import requests

import pandas as pd
```

```python
# attach link to a variable
```

```python
url = 'https://books.toscrape.com/catalogue/page-1.html'

link = requests.get(url)

print(link)
```

```python
# using BeautifulSoup to transform the data into html form
```

```python
soup = BeautifulSoup(link.text, 'html')
```

```python
# extrat the order list 'ol'from the data and assign to a variable
```

```python
ol = soup.find('ol')
```

```python
ol
```

```python
# the variables we need are all inside the 'article' tag under 'ol',
```

```python
articles = ol.find_all('article', class_='product_pod')
```

```python
articles
```

```python
# loop through all the articles and bring out all the values we intend to extract
# step we loop throught 'articles',
#       under attribute we create a variable called image in which we use to find the tag 'img'
#      create another variable called title and assign the image attribute 'image.attrs' and the tag we intend to use['alt']
#       the ['alt']tag is where the book title is
# 
```

```python
books = []                         #create an empty list
for article in articles:
    image = article.find('img')    #<img 
    title = image.attrs['alt']     #title is embedded inside alt
    star = article.find('p')       # star rating is within 'class' which is inside <p tag
    star = star['class'][1]        #'1' to extract only index 1=(2)
    price = article.find('p', class_='price_color').text #price is inside the <p tag under the class name which is 'price color' and the .text to extract the text only
    price = price[1:]
    price = float(price[1:]) # 1: to extract only the numbers and convert to a float type
    books.append([title,price,star])       #.append to add values into the list we created under the 'book' variable
```

```python
books
```

```python
df = pd.DataFrame(books, columns=['Title', 'Price', 'Star Rating'])   #convert into a dataframe
```
